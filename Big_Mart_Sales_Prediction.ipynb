# 3. Handling Missing Values

Mean and Mode methods of handling missing values

Replacing The Missing Values in Item_Weight Using Mean

```python
# Mean value of Item_Weight
df['Item_Weight'].mean()
```

```python
# Filling the missing values in Item_Weight column with the mean value
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].mean())
```

```python
df.isnull().sum()
```

Replacing Missing Values in Outlet_Size using Mode

```python
# Create a pivot table to find the most frequent (mode) Outlet_Size for each Outlet_Type
mode_of_Outlet_Size = df.pivot_table(
    values='Outlet_Size',                    # The column we want to analyze
    columns='Outlet_Type',                   # Group data by the type of outlet
    aggfunc=(lambda x: x.mode()[0])          # Apply a function that returns the most common (mode) value
)

# Display the resulting pivot table showing the mode Outlet_Size for every Outlet_Type
print(mode_of_Outlet_Size)
```

```python
missing_values  = df['Outlet_Size'].isnull()
```

```python
print(missing_values)
```

```python
# Finding the missing values
df.loc[missing_values, 'Outlet_Size'] = df.loc[missing_values, 'Outlet_Type'].apply(lambda x: mode_of_Outlet_Size[x])
```

# 7. Label Encoding

```python
encoder = LabelEncoder()
```

```python
# Transforming all the seven Categoricat Columns
df['Item_Identifier'] = encoder.fit_transform(df['Item_Identifier'])
df['Item_Fat_Content'] = encoder.fit_transform(df['Item_Fat_Content'])
df['Item_Type'] = encoder.fit_transform(df['Item_Type'])
df['Outlet_Identifier'] = encoder.fit_transform(df['Outlet_Identifier'])
df['Outlet_Size'] = encoder.fit_transform(df['Outlet_Size'])
df['Outlet_Location_Type'] = encoder.fit_transform(df['Outlet_Location_Type'])
df['Outlet_Type'] = encoder.fit_transform(df['Outlet_Type'])
```

```python
df.head()
```

# 8. Splitting Our Data into Features and Targets

```python
X = df.drop(columns = 'Item_Outlet_Sales', axis = 1)
Y = df['Item_Outlet_Sales']
```

```python
X
```

```python
Y
```

# 9. Splitting Our Data Into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# 10. Training Our Model

```python
model = XGBRegressor()
```

# 11. Fitting our model

```python
model.fit(X_train, Y_train)
```

# 12. Model Evaluation

```python
# Prediction on Training Data
training_data_prediction = model.predict(X_train)
```

```python
# R Squared Value
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R Squared Value: ', r2_train)
```

```python
# Prediction on Testing Data
test_data_prediction = model.predict(X_test)
```

```python
# R squared Value
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R Squared Value: ', r2_test)
```
